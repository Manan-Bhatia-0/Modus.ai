```python
import text2emotion as te
import nltk
import pandas as pd
```

```python
entries = pd.read_csv('~/Desktop/entries.csv')  
```

```python
entries
```

```python
entries['date'] = pd.to_datetime(entries['date'])
entries['entry'] = entries['entry'].astype(str)
```

```python
entries.dtypes
```

```python
entries
```

```python
#tokenizing paragraph into list of sentences
from nltk.tokenize import sent_tokenize
import string
entries['sent_tokenized'] = ''
for idx in entries.index:
    #print("line: " + row['entry'])
    entries['sent_tokenized'][idx] = nltk.tokenize.sent_tokenize(entries.loc[idx]['entry'])
```

```python
entries
```

```python
entries.loc[1][2]
```

```python
entries['t2e_paragraph'] = ''
#T2E on entire paragraph
for idx in entries.index:
    mood = te.get_emotion(entries.loc[idx]['entry'])
    entries['t2e_paragraph'][idx] = mood
```

```python
entries
```

```python
entries['t2e_sent_list'] = ''
#T2E on sentences per entry
for idx in entries.index:
    sentence_list = entries.loc[idx]['sent_tokenized']
    mood_list = []
    for i in sentence_list:
        mood = te.get_emotion(i)
        mood_list.append(mood)
    entries['t2e_sent_list'][idx] = mood_list
```

```python
entries
```

```python
entries.to_csv('entries_t2e_experiment.csv')
```
