```python
import re
def retrieve_data_from_url(url):
    pattern = re.compile(r"[\w]{8}")
    list_8 = []
    dict_8 = {}
    r= urllib.request.urlopen(url).read()
    soup = BeautifulSoup(r, 'lxml')
    things = soup.find_all('p')
    lst = []

    for i in things:
        i_strip = i.text.strip()
        i_strip = re.sub('[\W_]+'," ", i_strip)
        split_text = i_strip.split(" ")
        lst = lst + split_text
    return lst
```

```python
lst = retrieve_data_from_url('https://en.wikipedia.org/wiki/Natural_language_processing')
```
